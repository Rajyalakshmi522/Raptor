```python
np.random.seed(6)
a=np.random.randint(10,200,50)
a
# Create a new array 
#a>50 and a<150 -good
#a<50 - bad
#a>150 - excellent
```

```python
l=[]
for i in a:
    if i>=50 and i<=150:
        l.append('good')
    elif(i<50):
        l.append('bad')
    elif(i>150):
        l.append('Excellent')
print(np.array(l))
```

```python
def fn(n):
    if n<50:
        return 'bad'
    elif n>=50 and n<=150:
        return 'good'
    elif n>150:
        return 'Excellent'
fn(200)
```

```python
num=np.vectorize(fn)
```

```python
num(a)
```
