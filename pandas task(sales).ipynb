### How much sales happening in each customer segment

```python
data.Customer_Segment.nunique()
```

```python
data.Customer_Segment.value_counts()
```

```python
(data['Sales'][data.Customer_Segment=='CORPORATE']).sum()
```

```python
segments=data.Customer_Segment.value_counts().index
segments
```

```python
Sales_Segments=[]
for segment in segments:
    Sales_Segments.append(
        (data['Sales'][data.Customer_Segment==segment]).sum())
Sales_Segments
```

### Create a pandas series with values as sum in each category and index as segment name

```python
Sales_Segments=pd.Series(Sales_Segments,segments)
Sales_Segments
```

### Findout the segment name where we have the highest sales

```python
Sales_Segments.idxmax()
```

```python
# Find the average sales in each customer segment
Sales_Segments=[]
for segment in segments:
    Sales_Segments.append(
        (data['Sales'][data.Customer_Segment==segment]).mean())
Sales_Segments
```

Define a function
- input is a string
- return a value must be 0,1,2,3 or 4

```python
def str_to_num(x):
    if x=='NOT SPECIFIED':
        return 0
    elif x=='LOW':
        return 1
    elif x=='MEDIUM':
        return 2
    elif x=='HIGH':
        return 3
    elif x=='CRITICAL':
        return 4
```

```python
data['Order_Priority_Number']=data['Order_Priority'].apply(str_to_num)
data['Order_Priority_Number'].value_counts()
```
