```python
l.sort()
print(l)
```

```python
for i in l:
    dic={'student_id':i}
    li=[]
    for j in data:
        if(i==j['student_id']):
            li.append(j['attendance'])
        dic['attendance']=li
           
    print(dic)
    
```
