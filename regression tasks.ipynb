#Model Generlization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
help(make_regression)
```

```python
x,y=make_regression(1000,1,n_targets=1,random_state=34)
```

```python
x
```

```python
print(x.shape)
print(type(x))
print(x.ndim)
print(y.shape)
print(type(y))
print(y.ndim)
```

```python
from sklearn.linear_model import LinearRegression
#Create an istance
LR=LinearRegression()
#train
LR.fit(x,y)
#predict with train data
y_pred=LR.predict(x)
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
help(LinearRegression)
```

```python
#create user define function name mse
#inputs y_pred,y
#output is mean squared Error
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_pred,y))
```

```python
LR.predict(np.array([[-1.5],[1.5],[0.344]]))
```

```python
-1.7*LR.coef_[0]+LR.intercept_
```
